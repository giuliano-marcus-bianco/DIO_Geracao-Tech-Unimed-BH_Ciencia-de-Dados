```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# Criar o data frame a partir de um csv // df = data frame
df = pd.read_csv('/content/drive/MyDrive/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Gapminder.csv', error_bad_lines=False, sep=';')
```

```python
# Visualizar as 5 primeiras linhas
df.head()
```

```python
# Renomear colunas
df = df.rename(columns={'country':'Pais', 'continent':'Continente', 'year':'Ano', 'lifeExp':'Expectativa de vida', 'pop':'Populacao', 'gdpPercap':'PIB per capta'})
```

```python
df.head(10)
```

```python
# Retornar o nome das colunas
df.columns
```

```python
# Retorna o tipo de dado de cada coluna // objetc = string
df.dtypes
```

```python
# Retornar últimas linhas do banco de dados
df.tail(15)
```

```python
# Retorna informações estatítsticas do banco de dados
df.describe()
```

```python
# Retornar valores únicos da coluna 'Continente'
df['Continente'].unique()
```

```python
# Retornar dados apenas do 'Continente' = 'Oceania'
Oceania = df.loc[df['Continente'] == 'Oceania']
Oceania.head()
```

```python
# Confirmar que a variável Oceania possui apenas dados 'Oceania' na coluna 'Continente'
Oceania['Continente'].unique()
```

```python
# Agrupar dados pela coluna 'Continente' e retornar a quantidade de valores distintos na coluna 'Pais' // group by
df.groupby('Continente')['Pais'].nunique()
```

```python
# Agrupar dados pela coluna 'Ano' e retornar a média dos valores na coluna 'Expectativa de vida'
df.groupby('Ano')['Expectativa de vida'].mean()
```

```python
# Retornar a média da coluna 'PIB per capta'
df['PIB per capta'].mean()
```

```python
# Retornar a soma da coluna 'PIB per capta'
df['PIB per capta'].sum()
```
