# **Visualização de dados**

```python
# Importando a biblioteca Pandas
import pandas as pd
```

```python
# Criar variável de data frame a partir dos arquivos .xlsx
dfAracaju = pd.read_excel('Aracaju.xlsx')
dfFortaleza = pd.read_excel('Fortaleza.xlsx')
dfNatal = pd.read_excel('Natal.xlsx')
dfRecife = pd.read_excel('Recife.xlsx')
dfSalvador = pd.read_excel('Salvador.xlsx')
```

```python
# Juntar (concatenar) todos os data frames em um só
df = pd.concat([dfAracaju, dfFortaleza, dfNatal, dfRecife, dfSalvador])
```

```python
# Criar coluna "Receita", multiplicando o valor na coluna "Vendas" pelo valor na coluna "Qtde"
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

```python
# Contar quantos registros tem para cada valor na coluna "LojaID"
# Ou seja, retorna o número de vendas que cada Loja realizou
df['LojaID'].value_counts(ascending=False)
```

```python
# Plotar um gráfico de barras comparando as lojas pela quantidade de vendas de cada
df['LojaID'].value_counts(ascending=False).plot.bar()
```

```python
# Plotar um gráfico de barras HORIZONTAIS comparando as lojas pela quantidade de vendas de cada
df['LojaID'].value_counts(ascending=True).plot.barh()
```

```python
# Retirar a linha sobre o gráfico // use ; ao final do comando
df['LojaID'].value_counts(ascending=True).plot.barh();
```

```python
# Plotar um gráfico de pizza comparando o somatória da coluna "Receita" agrupado por ano
df.groupby(df['Data'].dt.year)['Receita'].sum().plot.pie();
```

```python
# Contar quantos registros tem para cada valor na coluna "Cidade"
# Ou seja, retorna o número de vendas em cada cidade
df['Cidade'].value_counts(ascending=False)
```

```python
# Importar a biblioteca "matplotlib.pyplot"
import matplotlib.pyplot as plt
```

```python
# Plotar gráfico de comparação de vendas entre as cidade com título
df['Cidade'].value_counts().plot.bar(title='Total de vendas por Cidade')
# Alterar eixos do gráfico
plt.xlabel('Cidade')
plt.ylabel('Total de Vendas');
```

```python
# Plotar gráfico de ccom a cor VERDE
df['Cidade'].value_counts().plot.bar(title='Total de vendas por Cidade', color="Green")
# Alterar eixos do gráfico
plt.xlabel('Cidade')
plt.ylabel('Total de Vendas');
```

```python
# Alterar o estilo do gráfico (https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
plt.style.use('Solarize_Light2')
```

```python
# Criar gráfico de linha com legenda comparando quantidade de produtos vendidos por mês
df.groupby(df['Data'].dt.month)['Qtde'].sum().plot(title='Total de produtos vendidos por mês')
plt.xlabel('Mês')
plt.ylabel('Total de produtos')
plt.legend();
```

```python
# Retornar total das vendas agrpado por mês
df.groupby(df['Data'].dt.month)['Qtde'].sum()
```

```python
# Filtrar pelo ano 2019
df_2019 = df[df['Data'].dt.year == 2019]
```

```python
# Plotar gráfico de total de vendas por mês no ano de 2019
# Adicionar marcador no gráfico // plot(marker="")
df_2019.groupby(df_2019['Data'].dt.month)['Qtde'].sum().plot(title='Total de produtos vendidos por mês',marker = 'o')
plt.xlabel('Mês')
plt.ylabel('Total de produtos')
plt.legend();
```

```python
# Plotar um gráfico histograma com a quantidade de produtos vendidos
# Mudar a cor para "olive" (https://matplotlib.org/stable/gallery/color/named_colors.html?highlight=named_colors)
plt.hist(df['Qtde'], color='olive');
```

```python
# Platar um gráfico de dispersão com quantidade de produtos vendidos pela receita
plt.scatter(x = df_2019['Data'].dt.day, y = df_2019['Receita']);
```

```python
# Salvar o gráfico como png
df_2019.groupby(df_2019['Data'].dt.month)['Qtde'].sum().plot(title='Total de produtos vendidos por mês',marker = 'o')
plt.xlabel('Mês')
plt.ylabel('Total de produtos')
plt.legend()
plt.savefig('Grafico QTDE x MES.png');
```
