# Estrutura de dados

```python
# Criando uma lista chamada animais
animais = [1, 2, 3]
animais
```

```python
# Atualizando a lista
animais = ['cachorro', 'gato', 123456, 3.14]
animais
```

```python
# Imprimir o primeiro elemento da lsita
animais[0]
```

```python
# Imprimindo o quarto elemento da lista
animais[3]
```

```python
# Substituindo primeiro elemento da lista
animais[0] = 'papagaio'
animais
```

```python
# Removendo gato da lista
animais.remove('gato')
animais
```

```python
# Sabendo o tamanho da lista
len(animais)
```

```python
# Sabendo se um elemento pertence à lista
'gato' in animais
```

```python
lista = [500, 30, 300, 80, 10]
```

```python
# Maior valor da lista
max(lista)
```

```python
# Menor valor da lista
min(lista)
```

```python
# Adicionar elemento à lista
animais.append(['leão', 'cachorro'])
```

```python
animais
```

```python
# Adicionar mais de um elemento à lista
animais.remove('cobra', 'rato')
```

```python
animais
```

```python
# Contanto o número de elementos em uma lista
animais.count('cachorro')
```

```python
animais.count('leão')
```

```python
# Ordernar lista do menor para o maior
lista.sort()
```

# Tuplas

```python
# As tuplas usam parênteses como sitaxe
tp = ('Banana', 'Maçã', 10, 50)
```

```python
# Retornar primeiro elemento (slice)
tp[0]
```

```python
# Diferente das listas, as tuplas são imutáveis. Ou seja, não podemos alterar seus elementos.
tp[0] = 'Laranja'
```

```python
tp.count('Maçã')
```

```python
tp[0:2]
```

# Dicionários

```python
# Para criar um dicionário, utilizamos as {} como sintaxe
dc = {'Maçã':20, 'Banana':10, 'Laranja':15, 'Uva':5} # Dicionários trabalham com o conceito de chave e valor
```

```python
dc
```

```python
# Acessando o valor de um dicionário através da chave
dc['Maçã']
```

```python
# Atualizando o valor de uma chave
dc['Maçã'] = 25
dc
```

```python
# Retornar as chaves do dicionário
dc.keys()
```

```python
# Retornar valores do dicionário
dc.values()
```

```python
# Verifica se já há existe chave no dicionário. Caso não exista, insere chave e valor.
dc.setdefault('Limão', 22)
```

```python
dc
```
