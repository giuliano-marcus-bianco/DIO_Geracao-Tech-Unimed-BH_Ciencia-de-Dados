```python
# Importar as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# Criar data frame
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
# Visualizar 5 primeiras linhas
df.head()
```

```python
# Verificar quantidade de linhas e colunas
df.shape
```

```python
# Verificar os tipos de dados
df.dtypes
```

```python
# Qual a receita total?
df['Valor Venda'].sum()
```

```python
# Qual o custo total?
df['Custo Unitário'].mul(df['Quantidade']).sum()
```

```python
# Criar a coluna "Custo"
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
# Qual foi o lucro total?
df['Valor Venda'].sum() - df['Custo Unitário'].mul(df['Quantidade']).sum()
```

```python
# Criar coluna "Lucro"
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
# Verificar alterações na tabela
df.head()
```

```python
# Criar coluna "Tempo de envio"
df['Tempo de envio'] = df['Data Envio'] - df['Data Venda']
```

```python
# Verificar alterações na tabela
df.head()
```

```python
# Qual o tempo médio de envio?
df['Tempo de envio'].mean()
```

**Transformar a coluna "Tempo de envio" em tipo numérico antes de consultar a média de tempo de envio agrupado por Marca**

```python
# Transormar a coluna "Tempo de envio" em tipo numérico
df['Tempo de envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df['Tempo de envio'].dtype
```

```python
# Retornar média tempo de envio agrupado por marca
df.groupby('Marca')['Tempo de envio'].mean()
```

**Tratar os valores ausentes na base de dados**

```python
# Verificar se existem valores null
df.isnull().sum()
```

**Retornar o Lucro agrupado por Ano e por Marca**

```python
# Retornar Lucro agrupando por Ano e por Marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Melhorar a visualização
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Retornar Lucro agrupando por Ano e por Marca com a visualização melhorada
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Criar uma nova tabela com o resultado obtido // reset_index()
lucro_ano_marca = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
```

```python
# Verificar como ficou a tabela lucro_ano_marca
lucro_ano_marca
```

```python
# Qual o total de produtos vendidos agrupado por produtos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Plotar em um gráfico de barras horizontais o total de produtos vendidos agrupado por produtos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
# Plotar em um gráfico de barras o lucro por ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro por ano')
plt.xlabel('Ano')
plt.ylabel('Lucro');
```

```python
# Criar tabela com vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
# Visualizar tabela com as vedas de 2009
df_2009
```

```python
# Plotar um gráfico de Lucro por mês das vendas de 2009
df_2009.groupby(df['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro por mês em 2009')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
# Plotar um gráfico de Lucro por Marca das vendas de 2009 com os ticks de x na posição horizontal
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro por marca em 2009')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
# Plotar um gráfico de Lucro por Classe das vendas de 2009 com os ticks de x na posição horizontal
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro por classe em 2009')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
